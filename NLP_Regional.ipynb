```python
# Descargamos lista de Stopwords en Español

nltk.download('stopwords')
stop_words=set(stopwords.words('spanish'))
```

```python
# Definimos función para remover puntuación

def remove_punctuations(text):
  translator=str.maketrans('','', string.punctuation)
  return text.translate(translator)
```

```python
# Función para eliminar tildes

def remove_accents(text):
  return unidecode(text)
```

```python
# Importamos Departamental

doc1=docx.Document('/content/PDF_Soportes.docx')
document2=''.join([para.text for para in doc1.paragraphs])
```

```python
# Removemos puntuación y tildes

document1=remove_punctuations(document2)
document1=remove_accents(document1)
```

```python
# Tokenizamos Documento

document1_tokens=nltk.word_tokenize(document1.lower())
```

```python
# Removemos Stop Words

document1_filtered=[word for word in document1_tokens if word not in stop_words]
```

```python
# Calculamos Frecuencias de Palabras

document1_freq=FreqDist(document1_filtered)
```

```python
# Imprimimos Palabras más Frecuentes

print("Plan Desarrollo Departamental", document1_freq.most_common(20))
```

```python
# Gráfico de Frecuencias más comunes

plt.figure(figsize=(10,5))
most_common_words1=[word for word, freq in document1_freq.most_common(20)]
freqs1 = [freq for word, freq in document1_freq.most_common(20)]
plt.bar(most_common_words1, freqs1)
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Document 1 - Most Common Words')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Gráficos de Dispersión Léxica

from matplotlib.cm import get_cmap

# Gráficos de Dispersión Léxica

target_words = ['bioeconomia','bioetica','bioinformatica','bioingenieria',
                'bionanotecnologia','biotecnologia','capital','ciberseguridad','datos','cientificas','cluster','comision','comite','compra','circular','digital','ecosistema','emprendimiento',
                'energia','experimentacion','financiacion','cuantica','innovacion','incubadora','index','indice','inteligencia','internet','investigacion','nanotecnologia'
                ,'intelectual','prototipado','redes','robotica','sandbox','tecnologias']

word_indices = {word: [i for i, w in enumerate(document1_filtered) if w == word] for word in target_words}

fig, ax = plt.subplots(figsize=(12, 6))

# Utilizamos mapa de colores

cmap = get_cmap('tab20')  # Colores distintos

for i, word in enumerate(target_words):
    indices = word_indices[word]
    ax.plot(indices, [i] * len(indices), marker='o', linestyle='', color=cmap(i/len(target_words)), label=word)

ax.set_yticks(range(len(target_words)))
ax.set_yticklabels(target_words)
ax.set_xlabel('Word Index')
ax.set_ylabel('Word')
ax.set_title('Lexical Dispersion Plot')

box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)

plt.show()
```

```python
import spacy
import docx
from nltk import Text
nlp=spacy.load('en_core_web_sm')
```

```python
def perform_ner(text):

    if not isinstance(text, str):
        text = " ".join(text)

    doc = nlp(text)

    entities = [(entity.text, entity.label_) for entity in doc.ents]

    return entities
```

```python
entities=perform_ner(document2)
print(entities)
```

```python
from collections import Counter
```

```python
# Graficamos Frecuencias

entity_counts=Counter([entity[1] for entity in entities])

plt.figure(figsize=(10, 6))
plt.bar(entity_counts.keys(), entity_counts.values(), color='skyblue')
plt.xlabel('Tipo de Entidad')
plt.ylabel('Frecuencia')
plt.title('Resultados NER')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
